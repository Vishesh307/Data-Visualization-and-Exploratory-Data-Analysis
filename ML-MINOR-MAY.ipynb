Objective of this Analysis:Â¶
To understand the how the student's performance (test scores) is affected by the other variables (Gender, Ethnicity, Parental level of education, Lunch, Test preparation course).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
#Visualising correlation using heatmap
sns.heatmap(cor)

```

```python
#checking impact of gender on the results of students on the basis of their test preparation by boxplot visualisation

plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'gender', y = 'writing score',hue='test preparation course',data = df)
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'gender', y = 'reading score',hue='test preparation course',data = df)
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x = 'gender', y = 'math score', hue='test preparation course',data = df)
```

```python
#from the above boxplot that for reading and writing scores the mean of scores of Female candidates(with or without test preparation) are above male candidates
#Also test preparation also affects the scores of the candidates and usually increases.
```

```python
#Now we have to check the impact of students education based on their parents education
sns.stripplot(x = "parental level of education", y = "math score",data = df,size=3)
plt.figure(figsize=(10,5))
plt.show()
sns.stripplot(x = "parental level of education", y = "reading score",data = df,size=3)
plt.figure(figsize=(10,5))
plt.show()
sns.stripplot(x = "parental level of education", y = "writing score",data = df,size=3)
plt.figure(figsize=(10,5))
plt.show()
```

```python
#from the above visualisation we get that scores in 3 exams are not dependent on the parental education since they have the similar distribution of scores 
#therefore it implies that parent education is an insignificant variable
```

```python
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x='race/ethnicity',y='math score',data=df,hue='lunch')
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x='race/ethnicity',y='reading score',data=df,hue='lunch')
plt.figure(figsize=(20, 12))
plt.subplot(2,3,1)
sns.boxplot(x='race/ethnicity',y='writing score',data=df,hue='lunch')
```

```python
#from the above visualisation we see that the race/ethnicity does not much effect the scores since the mean of scores of all the ethnic groups are almost equal
#But lunch variable has an effect on the scores as the diff btw the means of standard lunch is much more compared to free/reduced lunch
```

```python
#from the above analysis we get that the variables-race/ethnicity and parent level of education are the least significant in deciding the scores of the candidate
```
